```python
from pickle import STOP
text = open('batman.txt',"r").read()
stopwords = set(STOPWORDS)

wc = WordCloud(background_color="white", max_font_size = 50, max_words = 100, stopwords = stopwords, height = 1000, width = 1500)
wc.generate(text)

plt.imshow(wc,interpolation="bilinear")
plt.axis('off')
plt.show()

```

```python
custom_mask = np.array(Image.open("cloud.png"))

wc = WordCloud(background_color="white",stopwords = stopwords, mask = custom_mask)
wc.generate(text)

plt.imshow(wc,interpolation="bilinear")
plt.axis('off')
plt.show()
```

```python
custom_mask = np.array(Image.open("batman.png"))

wc = WordCloud(background_color="white",stopwords = stopwords, mask = custom_mask)
wc.generate(text)

plt.imshow(wc,interpolation="bilinear")
plt.axis('off')
plt.show()
```

```python
wc = WordCloud(background_color="white",stopwords = stopwords, mask = custom_mask)

wc.generate(text)
image_colors = ImageColorGenerator(custom_mask)
wc.recolor(color_func = image_colors)

plt.imshow(wc,interpolation="bilinear")
plt.axis('off')
plt.show()
```

```python
custom_mask = np.array(Image.open("batman_yellow.png"))

wc = WordCloud(background_color="yellow",stopwords = stopwords, mask = custom_mask)

wc.generate(text)
image_colors = ImageColorGenerator(custom_mask)
wc.recolor(color_func = image_colors)

plt.imshow(wc,interpolation="bilinear")
plt.axis('off')
plt.show()
```

```python
wc = WordCloud(background_color=(255, 242, 0),
               max_words = 1000, max_font_size = 30,
               stopwords = stopwords, mask = custom_mask)

wc.generate(text)
image_colors = ImageColorGenerator(custom_mask)
wc.recolor(color_func = image_colors)

plt.imshow(wc,interpolation="bilinear")
plt.axis('off')
plt.show()
```

```python
custom_mask = np.array(Image.open("batman.png"))
wc = WordCloud(background_color=(255, 242, 0),
               stopwords = stopwords, mask = custom_mask,
               contour_width= 5,
               contour_color= "black")

wc.generate(text)
image_colors = ImageColorGenerator(custom_mask)
wc.recolor(color_func = image_colors)

plt.imshow(wc,interpolation="bilinear")
plt.axis('off')
plt.show()

# or without background

wc = WordCloud(background_color="white",
               stopwords = stopwords, mask = custom_mask,
               contour_width= 5,
               contour_color= "black")

wc.generate(text)
image_colors = ImageColorGenerator(custom_mask)
wc.recolor(color_func = image_colors)

plt.imshow(wc,interpolation="bilinear")
plt.axis('off')
plt.show()
```
